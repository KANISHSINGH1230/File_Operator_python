```python
import os
from datetime import datetime

class ADD_Entries :

    def __init__ (self , filename = 'journal.py'):
        self.filename = filename

    def Add_entry (self) :
        try :
            entry = input("\n\n\t    Enter your journal entry : ")
            time = datetime.now().strftime("%d-%m-%y  %H:%M:%S")

            with open('journal.py' , 'a') as a :
                a.write(F'\n\n {time} \n {entry}')

        except FileNotFoundError :
            print("\n\n This file is not Found !...")
        else :
            print("\n\n The Entry is successfully Added !...")


class View_All_Entries (ADD_Entries) :

    def view_entry (self) :
        try :
            with open('journal.py' , 'r') as b :
                print(b.read())

        except FileNotFoundError :
            print("\n\n This file is not Found !...")
        else :
            print("\n\n These all are the Entries of the Journal !...")


class Search_An_Entries (View_All_Entries) :

    def search_entry (self) :
        try :
            keyword = input("\n\n\t    Enter the Keyword or date to search : ")

            a = False
            with open('journal.py' , 'r') as c :
                for i in c :
                    if ( keyword in i.lower() ) :
                        print("\n\n" , i.strip())
                        a = True

        except FileNotFoundError :
            print("\n\n This file is not Found !...")


class Delete_All_Entries (Search_An_Entries) :

    def delete_entry (self) :
        try :
            os.remove(self.filename)

        except FileNotFoundError :
            print("\n\n This file is not Found !...")
        else :
            print("\n\n The Entry is successfully Deleted !...")


class Call (Delete_All_Entries) :

    def menu (self) :
        while True :
            try :
                menu = input("""\n\n Select an option :  \n
            1. Add a Entry
            2. View all entries
            3. Search an entry
            4. Delete an entries
            5. Exit  \n
            Enter your choice : """)

                if ( menu == '1' ) :
                    self.Add_entry()
                    print("\n\n" , "*" * 70)
                elif ( menu == '2' ) :
                    self.view_entry()
                    print("\n\n" , "*" * 70)
                elif ( menu == '3' ) :
                    self.search_entry()
                    print("\n\n" , "*" * 70)
                elif ( menu == '4' ) :
                    self.delete_entry()
                    print("\n\n" , "*" * 70)
                elif ( menu == '5' ) :
                    break
                else :
                    print("\nPlease enter the valid input !...")

            except Exception as z :
                print(f"Error : {z}")


obj = Call()
obj.menu()
```
